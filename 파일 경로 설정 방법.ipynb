* ../../는 두 단계 위의 상위 폴더

## 여러 파일을 한번에 로드하는 방법

```python
# data/store 폴더의 csv 파일에 대한 경로 확인
file_csv = glob("data/store/*.csv")
file_csv
```

```python
pd.read_csv(file_csv[0], low_memory=False)
```

```python
file_list = []
# 반복문을 이용해 파일 한꺼번에 불러오기
for file_csv_name in file_csv:
    print(file_csv_name)
    temp = pd.read_csv(file_csv_name, low_memory=False)
    file_list.append(temp)
```

```python
file_list[0]
```

```python
len(file_list) # 17개의 파일
```

```python
df = pd.concat(file_list) # 콜롬 명이 같으면, 하나의 데이터프레임으로 합치기
df.shape
```
